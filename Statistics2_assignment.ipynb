**1. What is hypothesis testing in statistics?** 
- Hypothesis testing is a statistical method used to make decisions about a population based on sample data.It is a process to determine whether the assumption we make about a population parameter is correct or not.
- In hypothesis testing, we start by making two hypothesis:
- **Null Hypothesis(H‚ÇÄ):** It represents the default assumption that there is no effect, no difference, or no relationship.
- **Alternate Hypothesis(H‚ÇÅ or Ha):** It represents what we want to prove that there is an effect, a difference or a relationship.
- e.g.: Suppose a medicine claims to cure a disease in 80% of cases. We can perform hypothesis testing to check whether this claim is valid based on sample data from patients.Thus, hypothesis testing helps researchers and decision-makers conclude whether there is enough evidence in a sample to support a particular belief about a population.

**2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**
- In statistics, when we perform hypothesis testing, we make two competing statements about a population parameter: the null hypothesis and the alternative hypothesis.

- **Null Hypothesis (H‚ÇÄ):**
The null hypothesis is the default or initial assumption that there is no effect, no difference, or no relationship between variables. It represents the status quo or a statement of no change.
We assume the null hypothesis is true until we find strong evidence against it.

- **Alternative Hypothesis (H‚ÇÅ or Ha):**
The alternative hypothesis is what we want to prove. It suggests that there is an effect, a difference, or a relationship between variables. It is accepted if the evidence is strong enough to reject the null hypothesis.

-**Example:**
- H‚ÇÄ: A new drug has no effect on curing a disease.
- H‚ÇÅ: A new drug does have an effect on curing a disease.

In hypothesis testing, we either reject the null hypothesis or fail to reject it. We never "accept" the null hypothesis; we only say there is not enough evidence against it.

**3. What is the significance level in hypothesis testing, and why is it important?** 
- The significance level, often denoted by Œ± (alpha), is a threshold used in hypothesis testing to decide whether to reject the null hypothesis.
- It represents the probability of making a Type I error, which means rejecting the null hypothesis when it is actually true.

- The significance level is usually set before conducting the test. Common values are:

- 0.05 (5%)

- 0.01 (1%)

- 0.10 (10%)

Interpretation:
A significance level of 0.05 means that there is a 5% risk of concluding that a difference exists when there is no actual difference.
- Example:
If Œ± = 0.05, and the p-value from the test is 0.03, then 0.03 < 0.05, so we reject the null hypothesis and conclude that the results are statistically significant.

**4. What does a P-value represent in hypothesis testing?**
- In hypothesis testing, the P-value(probability value) represents the probability of observing the sample data or something more extreme, assuming that the null hypothesis  is true.
- A low P-value (typically ‚â§ 0.05) suggests that the observed data is unlikely under the null hypothesis ‚Äî so we reject the null hypothesis.
- A high P-value (> 0.05) means the data is likely under the null hypothesis ‚Äî so we fail to reject the null hypothesis.
- example: A company claims that their battery lasts 10 hours on average. You're a researcher and suspect it lasts less than 10 hours, so you perform a hypothesis test.

- Null Hypothesis (H‚ÇÄ): The battery lasts 10 hours (Œº = 10).

- Alternative Hypothesis (H‚ÇÅ): The battery lasts less than 10 hours (Œº < 10).

- You test 30 batteries and find the average life is 9.6 hours.
After performing the statistical test, you get a P-value = 0.02.

**5. How do you interpret the P-value in hypothesis testing?**
- The P-value helps you decide whether to reject or fail to reject the null hypothesis.
- **P-value ‚â§ 0.05 ‚Üí Statistically significant**
‚Üí There is strong evidence against the null hypothesis
‚Üí Reject the null hypothesis

- **P-value > 0.05 ‚Üí Not statistically significant**
‚Üí There is weak evidence against the null hypothesis
‚Üí Fail to reject the null hypothesis

**6. What are Type 1 and Type 2 errors in hypothesis testing?**
- **Type 1 Error (False Positive):**
- You reject the null hypothesis (H‚ÇÄ) when it is actually true.

- It means you're seeing an effect that isn't really there.

- Example: A COVID test says you're positive, but you're actually healthy.

- Probability of Type 1 Error = Œ± (alpha) ‚Üí usually set at 0.05 (5%)

- **Type 2 Error (False Negative):**
- You fail to reject the null hypothesis (H‚ÇÄ) when it is actually false.It means you miss a real effect.

- Example: A COVID test says you're negative, but you're actually infected.

- Probability of Type 2 Error = Œ≤ (beta)

**11. What is the difference between a Z-test and a T-test?**
- The Z-test and T-test are both statistical methods used to determine whether there is a significant difference between means. They are often applied in hypothesis testing but are used in different scenarios based on sample size and the availability of population data.
- **Z-test:**
A Z-test is used when the population variance is known and the sample size is large (typically more than 30). It assumes that the data follows a normal distribution. The Z-test relies on the standard normal distribution (also known as the Z-distribution). It is commonly used to test hypotheses about population means or proportions.
- **T-test:**
A T-test is used when the population variance is unknown and the sample size is small (usually less than or equal to 30). It uses the t-distribution, which is similar to the normal distribution but has heavier tails. This accounts for the increased uncertainty in small samples. There are different types of T-tests, including the one-sample T-test, independent two-sample T-test, and paired sample T-test.

**12. What is the T-test, and how is it used in hypothesis testing?**

**Use in Hypothesis Testing:**

3. Determine the degrees of freedom:
Usually, df=n‚àí1

4. Find the critical value or p-value using a T-distribution table.

5. Compare the T-statistic to the critical value:

6. If ‚à£t‚à£> critical value or p-value < significance level (Œ±), reject the null hypothesis.Otherwise, fail to reject the null hypothesis

**14. What is a confidence interval, and how is it used to interpret statistical results?** 
- A confidence interval gives an estimated range of values which is likely to include the unknown population parameter. This range is calculated from the sample statistics and depends on the confidence level selected (commonly 90%, 95%, or 99%).
- **Formula:** CI = X +- (Z or t) * s/‚úìn
- Where:
1. X = sample mean
2. s = sample standard deviation
3. n = sample size
4. Z or ùë° = critical value depending on confidence level and sample size.
- **Interpret a Confidence Interval:**
If we construct a 95% confidence interval, it means that we are 95% confident that the true population mean lies within this interval. It does not mean there's a 95% chance that the population mean is in that interval ‚Äî rather, if we repeated the sampling many times, 95% of the calculated intervals would contain the true mean.
Example:
- Suppose we calculate a 95% confidence interval for the average height of a group of students as:  - - [167.5¬†cm,172.5¬†cm]
This means we are 95% confident that the true average height of all students lies between 167.5 cm and 172.5 cm.

**15. What is the margin of error, and how does it affect the confidence interval?**

- **Affects on the Confidence Interval:**

**16. How is Bayes' Theorem used in statistics, and what is its significance?**

**Use in Statistics:**

**19. What is the F-distribution, and when is it used in hypothesis testing?** 
- The F-distribution is a continuous, right-skewed probability distribution that arises from the ratio of two independent chi-square variables divided by their degrees of freedom. It is always positive and depends on two degrees of freedom: one for the numerator and one for the denominator.

In hypothesis testing, the F-distribution is mainly used in:

**1. ANOVA (Analysis of Variance)**: To test whether the means of three or more groups are equal by comparing between-group and within-group variances.

**2. Variance Comparison:** To test if two populations have equal variances.

**3. Regression Analysis:** To test the overall significance of a regression model.

**20. What is an ANOVA test, and what are its assumptions?** 
- ANOVA (Analysis of Variance) is a statistical test used to determine whether there are significant differences between the means of three or more independent groups. Instead of comparing means one by one (which increases error), ANOVA compares overall group variance to detect any mean differences.

Null Hypothesis (H0): All group means are equal.

Alternative Hypothesis (H1): At least one group mean is different.

**Assumptions of ANOVA:**
**1. Independence:** Observations within and between groups must be independent.

**2. Normality:** Data in each group should be approximately normally distributed.

**3. Homogeneity of Variances:** All groups should have roughly equal variances (also called homoscedasticity).

**21. What are the different types of ANOVA tests?** 
- There are several types of ANOVA (Analysis of Variance) tests, each used based on the number of factors (independent variables) and how data is collected:
1. One-Way ANOVA
Use: Compares the means of three or more groups based on one independent variable.

Example: Comparing average marks of students from three different classes.

2. Two-Way ANOVA
Use: Examines the effect of two independent variables on one dependent variable, and can also detect interaction effects.

Example: Studying how both teaching method and gender affect test scores.

3. Repeated Measures ANOVA
Use: Used when the same subjects are measured multiple times under different conditions or over time.

Example: Measuring blood pressure of patients before, during, and after treatment.

**22. What is the F-test, and how does it relate to hypothesis testing?** 
- The F-test is a statistical method used to compare the variances of two or more groups to determine if they are significantly different from each other. It is based on the F-distribution, which is the ratio of two independent chi-square variables divided by their respective degrees of freedom. The F-test is commonly used in Analysis of Variance (ANOVA) and regression analysis.
-The main purpose of the F-test is to evaluate whether the group means are significantly different by comparing the variability between the groups to the variability within the groups.
- Formula: F = Variance bw groups / Variance within groups
- A large F-value suggests a large difference between group means.
- A small F-value suggests the group means are similar.
- Relation to Hypothesis Testing:
- The F-test is directly used in hypothesis testing with the following hypotheses:

 Null Hypothesis (H‚ÇÄ): All group variances (or means, in the case of ANOVA) are equal.

 Alternative Hypothesis (H‚ÇÅ): At least one group variance (or mean) is different.

-If the calculated F-value is greater than the critical F-value from the F-distribution table, we reject the null hypothesis, indicating a statistically significant difference.

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np 
from scipy import stats 

sample_data = [72, 75, 78, 74, 77, 73, 76, 80, 79, 81] 
population_mean = 70 
population_std = 5 

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 1 - stats.norm.cdf(z_score) 

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis ‚Äî the sample mean is significantly greater than the population mean.")
else:
    print("Result: Fail to reject the null hypothesis ‚Äî no significant difference between sample and population mean.")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot. 
import numpy as np 
import scipy.stats as stats 
import matplotlib.pyplot as plt 

mu = 100 
sigma = 15 
x_bar = 108
n = 36
alpha = 0.5 

standard_error = sigma / np.sqrt(n)
z = (x_bar - mu)/ standard_error
z_critical = stats.norm.ppf(1 - alpha/2)

print(f"Z-score: {z:.2f}")
print(f"Critical Z-value: ¬±{z_critical:.2f}") 

x = np.linspace(-4,4,1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(10,5))
plt.plot(x,y, label='Standard Normal Distribution')

plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color = 'red', alpha = 0.3, label= 'Rejection Region')
plt.axvline(z, color='blue', linestyle = '--', label=f'Z = {z: .2f}')
plt.axvline(-z_critical, color='black', linestyle='--')
plt.axvline(z_critical, color='black', linestyle='--')

plt.title("Two-Tailed Z- Test: Rejection Regions")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.stats as stats 

def visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha): 
    se = sigma / np.sqrt(n)

    z_critical = stats.norm.ppf(1 - alpha)
    x_critical = mu0 + z_critical * se 

    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y_null = stats.norm.pdf(x, mu0, se) 
    y_alt = stats.norm.pdf(x, mu1, se)

    plt.figure(figsize=(12,6))
    plt.plot(x, y_null, label='Ho Distribution (Œº = {})'.format(mu0), color='blue')
    plt.plot(x,y_alt, label= 'H1 Distribution (Œº= {})'.format(mu1), color='green')
    plt.fill_between(x, y_null, where=(x > x_critical), color='red', alpha=0.3, label='Type 1 Error (Œ±)')
    plt.fill_between(x, y_null, where=(x < x_critical), color='orange', alpha=0.3, label='Type 2 Error (Œ≤)') 

    plt.axvline(x_critical, color='black', linestyle= '--', label=f'Critical Value = {x_critical:.2f}')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results. 
import numpy as np 
from scipy import stats 

group1 = [85, 90, 88, 75, 95, 79, 84]
group2 = [80, 78, 75, 70, 72, 68, 74] 

t_stat, p_value = stats.ttest_ind(group1, group2)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two groups.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the two groups.")
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance? 
import numpy as np 
import scipy.stats as stats 

def confidence_interval(sample, confidence= 0.95): 
    n = len(sample)
    mean = np.mean(sample)
    std_err = stats.sem(sample)
    t_crit = stats.t.ppf((1 + confidence) / 2, df = n -1)
    margin = t_crit * std_err 
    lower = mean - margin 
    upper = mean + margin
    return round(lower, 2), round(upper, 2)

data = [52, 55, 53, 60, 58, 57, 54, 56, 59, 61]
ci = confidence_interval(data)
print("95% Confidence Interval:", ci)
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data? 
import numpy as np 
import scipy.stats as stats 

def calculate_margin_of_error(sample, confidence=0.95): 
    n = len(sample) 
    std_err = stats.sem(sample)

    t_crit = stats.t.ppf((1+confidence) / 2, df = n - 1 )
    margin_error = t_crit * std_err 
    return round(margin_error, 2)

data = [52, 55, 53, 60, 58, 57, 54, 56, 59, 61]
moe = calculate_margin_of_error(data, confidence=0.95)
print("Margin of Error (95% confidence):", moe)
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process. 
def bayes_theorem(prior_A, b_givenA, b_notgivenA): 
    prior_not_A = 1 - prior_A 
    numerator = b_givenA * prior_A 
    denominator = numerator + (b_notgivenA * prior_not_A)
    posterior = numerator / denominator 
    return round(posterior, 4)

P_Disease = 0.01                 
P_Pos_given_Disease = 0.99        
P_Pos_given_NoDisease = 0.05    

result = bayes_theorem(P_Disease, P_Pos_given_Disease, P_Pos_given_NoDisease)

print("P(Disease | Positive Test) =", result)

```

```python
#16. Implement an F-test using Python to compare the variances of two random samples. 
import numpy as np 
from scipy.stats import f 

np.random.seed(42)
sample1 = np.random.normal(loc=10, scale=2, size=30)
sample2 = np.random.normal(loc=10, scale=3, size=30)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

if var1 > var2: 
    f_stat = var1 / var2 
    dfn = len(sample1) - 1
    dfd = len(sample2) - 1 
else:
    f_stat = var2 / var1
    dfn = len(sample2) - 1
    dfd = len(sample2) - 1

p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

alpha = 0.05
if p_value < alpha: 
    print("Reject the null hypothesis: Variances are significantly different.")
else: 
    print("Fail to reject the null hypothesis: Variances are not significantly different.")
    
    
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA. 
import numpy as np 
import scipy.stats as stats 
import matplotlib.pyplot as plt 
import seaborn as sns 

def check_anova_assumptions(*groups):
    all_data = np.concatenate(groups)
    labels = []
    for i, group in enumerate(groups): 
        labels += [f'Group {i+1}'] * len(group)
    print("=== Checking Normality (Shapiro-Wilk Test) ===")
    for i, group in enumerate(groups): 
        stat, p = stats.shapiro(group)
        print(f"Group {i+1}: W={stat:.4f}, p-value={p:.4f}")
    print("Checking Equal Variance")
    stat, p = stats.levene(*groups)
    print(f"Levene's Test: W= {stat:.4f}, p-value={p:.4f}")

    print("\n=== Checking Independence ===")
    print("‚ÑπÔ∏è Independence is assumed based on proper experimental design (e.g., random sampling). Cannot be tested directly from data.")

    plt.figure(figsize=(10,5))
    sns.boxplot(x=labels, y=all_data)
    plt.title("Boxplot of Groups (Visual check for spread)")
    plt.ylabel("Values")
    plt.grid(True) 
    plt.show()

    plt.figure(figsize=(10,5))
    for i, group in enumerate(groups):
        sns.kdeplot(group, label=f'Group {i+1}')
    plt.title("KDE Plot for Visual Normality check")
    plt.xlabel("Value")
    plt.legend()
    plt.grid(True)
    plt.show()

group1 = np.random.normal(loc=50, scale=5, size=30)  
group2 = np.random.normal(loc=52, scale=5, size=30)  
group3 = np.random.normal(loc=60, scale=5, size=30) 

check_anova_assumptions(group1,group2,group3)
```

```python
#20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results?
import pandas as pd 
import numpy as np 

np.random.seed(1)

fertilizer = ['A','B','C']
watering = ['Daily','Weekly']
data = []

for f in fertilizer: 
    for w in watering: 
        for _ in range(10): 
            growth = np.random.normal(loc=20 + 5 * fertilizer.index(f) - 2 * watering.index(w), scale=2)
            data.append([f,w,growth])
df = pd.DataFrame(data, columns=['Fertilizer','Watering','Growth'])

import statsmodels.api as sm 
from statsmodels.formula.api import ols 

model = ols("Growth ~ C(Fertilizer) * C(Watering)", data=df).fit()
annova_table = sm.stats.anova_lm(model, typ = 2)
print(annova_table)

import seaborn as sns 
import matplotlib.pyplot as plt 

sns.pointplot(data = df, x='Fertilizer', y='Growth', hue="Watering", dodge=True, markers=['o', 's'], capsize=.1, err_kws={'linewidth':1}, palette= 'colorblind')
plt.title("Interaction Plot: Fertilizer vs Watering on Growth")
plt.ylabel("Mean Plant Growth")
plt.grid(True)
plt.show()
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
import numpy as np 
from scipy.stats import chi2 

np.random.seed(42)
sample = np.random.normal(loc=50, scale=8, size=30)

sigma_0_squared = 64
n = len(sample)

sample_variance = np.var(sample, ddof=1)

chi_square_stat = (n-1) * sample_variance / sigma_0_squared
alpha = 0.05
df = n-1

chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1-alpha /2, df)

print(f"Sample Variance: {sample_variance:.2f}")
print(f"Chi-square Statistic: {chi_square_stat:.2f}")
print(f"Chi-square Critical Values: [{chi2_lower:.2f}, {chi2_upper:.2f}]")

if chi_square_stat < chi2_lower or chi_square_stat > chi2_upper:
    print("Reject the null hypothesis: The population variance is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in population variance.")

```

```python
#26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results. 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=100)
group2 = np.random.normal(loc=52, scale=15, size=100)

var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

F = var1 / var2 if var1 > var2 else var2 / var1 
df1 = len(group1) - 1 
df2 = len(group2) - 1 

from scipy.stats import f 
p_value = 2 *(1- f.cdf(F, df1, df2)) if var1 >var2 else 2 * f.cdf(F, df1, df2)

print(f"Sample Variance 1: {var1:.2f}")
print(f"Sample Variance 2: {var2:.2f}")
print(f"F-statistic: {F:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")

plt.figure(figsize=(10, 5))
sns.kdeplot(group1, label='Group 1', fill=True)
sns.kdeplot(group2, label='Group 2', fill=True)
plt.title("Distribution of Group 1 and Group 2")
plt.xlabel("Values")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
